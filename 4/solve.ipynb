# Part One


```python
with open("input.txt", "r") as file:
    input = file.read()

grid = [list(row) for row in input.split("\n") if len(row.strip()) > 0]

grid_height = len(grid)
grid_width = len(grid[0])

grid
```

```python
xmas_occurances = 0

for y in range(grid_height):
    for x in range(grid_width):
        if grid[y][x] != "X":
            continue

        # To the left
        if (
            x - 3 >= 0
            and grid[y][x - 1] == "M"
            and grid[y][x - 2] == "A"
            and grid[y][x - 3] == "S"
        ):
            xmas_occurances += 1
        # To left top diagonal
        if (
            x - 3 >= 0
            and y - 3 >= 0
            and grid[y - 1][x - 1] == "M"
            and grid[y - 2][x - 2] == "A"
            and grid[y - 3][x - 3] == "S"
        ):
            xmas_occurances += 1
        # To top
        if (
            y - 3 >= 0
            and grid[y - 1][x] == "M"
            and grid[y - 2][x] == "A"
            and grid[y - 3][x] == "S"
        ):
            xmas_occurances += 1
        # To top right diagonal
        if (
            y - 3 >= 0
            and x + 3 < grid_width
            and grid[y - 1][x + 1] == "M"
            and grid[y - 2][x + 2] == "A"
            and grid[y - 3][x + 3] == "S"
        ):
            xmas_occurances += 1
        # To the right
        if (
            x + 3 < grid_width
            and grid[y][x + 1] == "M"
            and grid[y][x + 2] == "A"
            and grid[y][x + 3] == "S"
        ):
            xmas_occurances += 1
        # To the bottom right
        if (
            y + 3 < grid_height
            and x + 3 < grid_width
            and grid[y + 1][x + 1] == "M"
            and grid[y + 2][x + 2] == "A"
            and grid[y + 3][x + 3] == "S"
        ):
            xmas_occurances += 1
        # To the bottom
        if (
            y + 3 < grid_height
            and grid[y + 1][x] == "M"
            and grid[y + 2][x] == "A"
            and grid[y + 3][x] == "S"
        ):
            xmas_occurances += 1
        # To the bottom left
        if (
            y + 3 < grid_height
            and x - 3 >= 0
            and grid[y + 1][x - 1] == "M"
            and grid[y + 2][x - 2] == "A"
            and grid[y + 3][x - 3] == "S"
        ):
            xmas_occurances += 1

xmas_occurances
```

# Part Two


```python
mas_in_x_shape_occurances = 0

for y in range(grid_height):
    for x in range(grid_width):
        if grid[y][x] != "A":
            continue

        # Bounds check
        if not 1 <= y < (grid_height - 1) or not 1 <= x < (grid_width - 1):
            continue

        top_left_bottom_right_mas = (
            grid[y - 1][x - 1] == "M" and grid[y + 1][x + 1] == "S"
        ) or (grid[y - 1][x - 1] == "S" and grid[y + 1][x + 1] == "M")

        bottom_left_top_right_mas = (
            grid[y + 1][x - 1] == "M" and grid[y - 1][x + 1] == "S"
        ) or (grid[y + 1][x - 1] == "S" and grid[y - 1][x + 1] == "M")

        if top_left_bottom_right_mas and bottom_left_top_right_mas:
            mas_in_x_shape_occurances += 1

mas_in_x_shape_occurances
```
