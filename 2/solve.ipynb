# Part One

Safe:

- The levels are either all increasing or all decreasing.
- Any two adjacent levels differ by at least one and at most three.


```python
with open("input.txt", "r") as file:
    input = file.read()


reports = [[int(num) for num in report.split()] for report in input.split("\n")]
reports = [report for report in reports if len(report) >= 1]

reports
```

```python
from typing import List

safe_reports = 0


def check_report(report: List[int]) -> int:
    fail_index = -1

    if len(report) <= 1:
        return fail_index

    sort_mode = "asc" if report[-1] > report[0] else "desc"

    safe = True

    for i in range(1, len(report)):
        prev = report[i - 1]
        curr = report[i]

        diff = curr - prev

        if sort_mode == "asc":
            safe = 1 <= diff <= 3
        else:
            safe = -3 <= diff <= -1

        if not safe:
            return i

    return fail_index


for report in reports:
    fail_index = check_report(report)

    if fail_index == -1:
        safe_reports += 1
    else:
        for i in range(len(report)):
            fail_index = check_report(
                [num for idx, num in enumerate(report) if idx != i]
            )
            if fail_index == -1:
                safe_reports += 1
                break


safe_reports
```
