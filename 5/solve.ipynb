# Part one

```python
with open("input.txt", "r") as file:
    input = file.read()

input
```

```python
import re

matches = re.findall("(\\d+\\|\\d+|(?:\\d+,?)+)", input)

rules = [tuple(int(el) for el in match.split("|")) for match in matches if "|" in match]
orderings = [
    [int(num) for num in match.split(",")] for match in matches if "," in match
]

rules, orderings
```

```python
from math import floor

middle_sum = 0

for ordering in orderings:
    correct = True

    for first, second in rules:
        try:
            first_index = ordering.index(first)
            second_index = ordering.index(second)

            if first_index > second_index:
                correct = False
                break
        except:
            pass

    if correct:
        middle_sum += ordering[floor(len(ordering) / 2)]

middle_sum
```

# Part two

```python
middle_sum = 0

for ordering in orderings:
    correct = True

    changed = True
    while changed:
        changed = False
        for first, second in rules:
            try:
                first_index = ordering.index(first)
                second_index = ordering.index(second)

                if first_index > second_index:
                    correct = False

                    ordering.insert(second_index, ordering.pop(first_index))
                    changed = True
            except:
                pass

    if not correct:
        middle_sum += ordering[floor(len(ordering) / 2)]

middle_sum
```
